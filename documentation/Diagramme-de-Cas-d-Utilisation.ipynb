```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from PIL import Image

# Create a new figure
fig, ax = plt.subplots(figsize=(14, 8))

# Define positions of elements with more spacing
elements = {
    "User": (0.1, 0.5),
    "GET /\nAfficher un message de bienvenue": (0.5, 0.8),
    "POST /predict\nPrdire une seule requte": (0.5, 0.5),
    "POST /predict_csv\nPrdire plusieurs requtes": (0.5, 0.2)
}

# Add elements to the plot
for text, (x, y) in elements.items():
    if "User" in text:
        box = patches.FancyBboxPatch((x, y), 0.2, 0.2, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightyellow')
    else:
        box = patches.FancyBboxPatch((x, y), 0.4, 0.15, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightblue')
    ax.add_patch(box)
    ax.text(x + 0.1 if "User" in text else x + 0.2, y + 0.1 if "User" in text else y + 0.075, text, horizontalalignment='center', verticalalignment='center', fontsize=10, fontweight='bold')

# Add a figure of a person to the User box
img_path = 'Basic_human_drawing.png'  # Update this path to your local image path
img = Image.open(img_path)
imagebox = OffsetImage(img, zoom=0.1)
ab = AnnotationBbox(imagebox, (0.2, 0.6), frameon=False)
ax.add_artist(ab)

# Add arrows for interactions
arrowprops = dict(facecolor='black', arrowstyle='->', lw=2)
interactions = [
    ("User", "GET /\nAfficher un message de bienvenue"),
    ("User", "POST /predict\nPrdire une seule requte"),
    ("User", "POST /predict_csv\nPrdire plusieurs requtes")
]

for start, end in interactions:
    start_x, start_y = elements[start]
    end_x, end_y = elements[end]
    ax.annotate('', xy=(end_x + 0.2, end_y + 0.075), xytext=(start_x + 0.2, start_y + 0.1), arrowprops=arrowprops)

# Set the limits and title
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('Diagramme de Cas d\'Utilisation', fontsize=15)

# Remove the axes
ax.axis('off')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots(figsize=(14, 8))

# Define positions of elements with more spacing
elements = {
    "User": (0.1, 0.5),
    "GET /\nAfficher un message de bienvenue": (0.5, 0.8),
    "POST /predict\nPrdire une seule requte": (0.5, 0.5),
    "POST /predict_csv\nPrdire plusieurs requtes": (0.5, 0.2)
}

# Add elements to the plot
boxes = {}
for text, (x, y) in elements.items():
    if "User" in text:
        box = patches.FancyBboxPatch((x, y), 0.2, 0.2, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightyellow')
        ax.add_patch(box)
        boxes[text] = box
        ax.text(x + 0.1, y - 0.01, text, horizontalalignment='center', verticalalignment='center', fontsize=10, fontweight='bold')
    else:
        box = patches.FancyBboxPatch((x, y), 0.4, 0.15, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightblue')
        ax.add_patch(box)
        boxes[text] = box
        ax.text(x + 0.2, y + 0.075, text, horizontalalignment='center', verticalalignment='center', fontsize=10, fontweight='bold')

# Add stick figure to the User box
stick_figure_x = 0.2
stick_figure_y = 0.6

# Head
head = plt.Circle((stick_figure_x, stick_figure_y + 0.05), 0.02, edgecolor='black', facecolor='none')
ax.add_patch(head)

# Body
ax.plot([stick_figure_x, stick_figure_x], [stick_figure_y, stick_figure_y + 0.03], color='black')

# Arms
ax.plot([stick_figure_x - 0.02, stick_figure_x + 0.02], [stick_figure_y + 0.01, stick_figure_y + 0.01], color='black')

# Legs
ax.plot([stick_figure_x, stick_figure_x - 0.02], [stick_figure_y, stick_figure_y - 0.03], color='black')
ax.plot([stick_figure_x, stick_figure_x + 0.02], [stick_figure_y, stick_figure_y - 0.03], color='black')

# Add arrows for interactions
arrowprops = dict(facecolor='black', arrowstyle='->', lw=2)
interactions = [
    ("User", "GET /\nAfficher un message de bienvenue"),
    ("User", "POST /predict\nPrdire une seule requte"),
    ("User", "POST /predict_csv\nPrdire plusieurs requtes")
]

for start, end in interactions:
    start_box = boxes[start]
    end_box = boxes[end]
    start_x, start_y = start_box.get_x() + start_box.get_width(), start_box.get_y() + start_box.get_height() / 2
    end_x, end_y = end_box.get_x(), end_box.get_y() + end_box.get_height() / 2
    ax.annotate('', xy=(end_x, end_y), xytext=(start_x, start_y), arrowprops=arrowprops)

# Set the limits and title
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('Diagramme de Cas d\'Utilisation', fontsize=15)

# Remove the axes
ax.axis('off')

# Show the plot
plt.show()

```
