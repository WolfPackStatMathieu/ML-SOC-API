```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots(figsize=(12, 8))

# Define positions of elements
api_box = patches.FancyBboxPatch((0.1, 0.6), 0.3, 0.2, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightblue')
s3_box = patches.FancyBboxPatch((0.6, 0.6), 0.3, 0.2, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightgreen')
mlflow_box = patches.FancyBboxPatch((0.1, 0.2), 0.3, 0.2, boxstyle="round,pad=0.05", edgecolor='black', facecolor='lightcoral')

# Add the elements to the plot
ax.add_patch(api_box)
ax.add_patch(s3_box)
ax.add_patch(mlflow_box)

# Add arrows
arrowprops = dict(facecolor='black', arrowstyle='->', lw=2)
ax.annotate('', xy=(0.4, 0.7), xytext=(0.6, 0.7), arrowprops=arrowprops)
ax.annotate('', xy=(0.25, 0.6), xytext=(0.25, 0.4), arrowprops=arrowprops)

# Add text
ax.text(0.1 + 0.15, 0.6 + 0.1, 'API FastAPI', horizontalalignment='center', verticalalignment='center', fontsize=12, fontweight='bold')
ax.text(0.6 + 0.15, 0.6 + 0.1, 'S3 Storage\n(MinIO)', horizontalalignment='center', verticalalignment='center', fontsize=12, fontweight='bold')
ax.text(0.1 + 0.15, 0.2 + 0.1, 'MLflow Server', horizontalalignment='center', verticalalignment='center', fontsize=12, fontweight='bold')

# Add labels to arrows
ax.text(0.5, 0.75, 'Load Preprocessing Pipeline\n(corresponding to model)', horizontalalignment='center', verticalalignment='center', fontsize=10)
ax.text(0.25, 0.5, 'Load Model in Production', horizontalalignment='center', verticalalignment='center', fontsize=10)

# Set the limits and title
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('Interaction between API, S3 Storage, and MLflow', fontsize=15)

# Remove the axes
ax.axis('off')

# Show the plot
plt.show()
```
